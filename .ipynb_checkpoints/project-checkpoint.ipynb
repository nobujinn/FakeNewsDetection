```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/News.csv',index_col=0)
data.head()
```

```python
data = data.drop(["title", "subject","date"], axis = 1)
```

Title, subject, date columns are not helpful for us in identification of news, so we drop them

```python
data.isnull().sum()
```

Check if any row is null

```python
data = data.sample(frac=1)
data.reset_index(inplace=True)
data.drop(["index"], axis=1, inplace=True)
```

We shuffle the data set to prevent the model to accidentally learn pattern from the order

```python
sns.countplot(data=data,
              x='class',
              order=data['class'].value_counts().index)
```

```python
from tqdm import tqdm
import re
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer
from wordcloud import WordCloud
```

```python
def preprocess_text(text_data):
    preprocessed_text = []
    
    for sentence in tqdm(text_data):
        sentence = re.sub(r'[^\w\s]', '', sentence)
        preprocessed_text.append(' '.join(token.lower()
                                  for token in str(sentence).split()
                                  if token not in stopwords.words('english')))

    return preprocessed_text
```

Function for preprocessing the text(remove any sings except the letters and numbers, make all lowercase, remove the stop words)

```python
preprocessed_review = preprocess_text(data['text'].values)
data['text'] = preprocessed_review
```

```python
# Real news
consolidated = ' '.join(
    word for word in data['text'][data['class'] == 1].astype(str))
wordCloud = WordCloud(width=1600,
                      height=800,
                      random_state=21,
                      max_font_size=110,
                      collocations=False)
plt.figure(figsize=(15, 10))
plt.imshow(wordCloud.generate(consolidated), interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# Fake news
consolidated = ' '.join(
    word for word in data['text'][data['class'] == 0].astype(str))
wordCloud = WordCloud(width=1600,
                      height=800,
                      random_state=21,
                      max_font_size=110,
                      collocations=False)
plt.figure(figsize=(15, 10))
plt.imshow(wordCloud.generate(consolidated), interpolation='bilinear')
plt.axis('off')
plt.show()
```

Visualization in a form of wordcloud for real and fake news
